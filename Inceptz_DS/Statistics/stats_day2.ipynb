```python
df = pd.read_csv('spotify.csv',index_col= "Date" , parse_dates =True)
```

```python
 
plt.figure(figsize= (17,6))
plt.title(label ="Trend of Spotify.CSV")
plt.ylabel('some numbers')
plt.grid(True)
sns.set_style('dark')
sns.lineplot(data =df )
```
