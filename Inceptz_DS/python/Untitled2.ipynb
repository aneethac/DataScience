```python
def square(x):
    return x*x

```

```python
df["sq Meal Price"] = list(map(lambda x: x*x,df["Meal Price"]))
```

```python
df["square Meal Price"] = square(df["Meal Price"])
```

```python
df["square1 Meal Price"] = df["Meal Price"].apply(square)
```
